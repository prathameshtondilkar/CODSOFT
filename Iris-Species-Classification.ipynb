```python
from sklearn.cluster import KMeans
```

<h4><i>Elbow Technique</i></h4>

```python
# Elbow Technique

sse_pk=[]
k_range=range(1,10)

for k in k_range:
    km=KMeans(n_clusters=k)
    km.fit(ds[['petal_length','petal_width']])
    sse_pk.append(km.inertia_)
```

```python
sse_pk
```

```python
plt.xlabel('k_range')
plt.ylabel('Sum of Squared Errors')
plt.plot(k_range,sse_pk)
```

<h4><i>KMeans Algorithm</i></h4>

```python
km = KMeans(n_clusters=3,random_state=5)
y_predicted=km.fit_predict(ds[['petal_length','petal_width']])
y_predicted
```

```python
ds['cluster']=y_predicted
ds.head(130)
```

```python
from sklearn.metrics import confusion_matrix
con_mat = confusion_matrix(ds.species , ds.cluster)
con_mat
```

<h4><i>Accuracy Measure</i></h4>
<h4><i>Confusion Matrix</i></h4>

```python
true_labels = ds.species
predicted_labels = ds.cluster

cm= confusion_matrix(true_labels ,predicted_labels)
class_labels = ['Setosa','Versicolor','Virginica']

plt.imshow(cm, interpolation = 'nearest' , cmap= plt.cm.Greens)
plt.title('Confusion Matrix')
plt.colorbar()
tick_marks = np.arange(len(class_labels))
plt.xticks(tick_marks , class_labels)
plt.yticks(tick_marks , class_labels)

for i  in range(len(class_labels)):
    for j in range(len(class_labels)):
        plt.text(j,i,str(cm[i][j]),ha='center',va='center',color='white')
        
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
